```python
#Create a 'TotalSales' column (Quantity * UnitPrice)
df['TotalSales'] = df['Quantity'] * df['UnitPrice']
product_sales = df.groupby("Description")["TotalSales"].sum().sort_values(ascending=False).head(10)
# Visualize sales per product using bar and line charts. Rohith John 24BAD100
plt.figure(figsize=(10,6))
plt.bar(product_sales.index, product_sales.values)
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.title("Top 10 Products by Sales")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
# 1. Convert InvoiceDate to datetime format
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
# 2. Now perform the resampling (using the corrected date column)
monthly_trend = df.set_index('InvoiceDate')['TotalSales'].resample('ME').sum()

# 3. Plotting (Fixed the typo: monthly_trend.values)
plt.figure(figsize=(10,5))
plt.plot(monthly_trend.index, monthly_trend.values, marker='o', linestyle='-', color='blue')
plt.title('Total Sales Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales ($)')
plt.grid(True)
plt.show()
```

```python
# Check for null values     Rohith JOHN 24BAD100
print("Missing Values per Column:")
missing_vals = df_housing.isnull().sum()
print(missing_vals)
# Visualizing missing values (Yellow lines would indicate missing data)
plt.figure(figsize=(10, 6))
sns.heatmap(df_housing.isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title('Missing Values Map (Solid color means no missing data)')
plt.show()
```

```python
# Compute correlation only on numerical columns (Price, Area, Bedrooms, etc.)
corr_matrix = df_housing.corr(numeric_only=True)
# Plot the Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Housing Features')
plt.show()
```

```python
# Scatter Plot: Price vs. Area Rohith John 24BAD100
plt.figure(figsize=(10, 6))
sns.scatterplot(x='area', y='price', data=df_housing, color='blue', alpha=0.6)
plt.title('House Price vs. Area')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df_bank = pd.read_csv('marketing_campaign.csv', sep=None, engine='python')
df_bank.columns = df_bank.columns.str.strip()
df_bank['Age'] = 2026 - df_bank['Year_Birth']
df_bank['TotalSpending'] = df_bank['MntWines'] + df_bank['MntFruits'] + df_bank['MntMeatProducts'] + df_bank['MntFishProducts'] + df_bank['MntSweetProducts'] + df_bank['MntGoldProds']
df_bank['Income'] = pd.to_numeric(df_bank['Income'], errors='coerce')
df_bank['Income'] = df_bank['Income'].fillna(df_bank['Income'].median())

# Visualizations for Age, Income, and Spending by Education Level 24BAD100 Rohith John
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sns.boxplot(y=df_bank['Age'], color='skyblue')
plt.title('Age Distribution')
plt.ylim(0, 100)
plt.subplot(1, 3, 2)
sns.boxplot(y=df_bank['Income'], color='lightgreen')
plt.title('Income Distribution')
plt.ylim(0, 200000)
plt.subplot(1, 3, 3)
sns.barplot(x='Education', y='TotalSpending', data=df_bank, palette='viridis', errorbar=None)
plt.title('Avg Spending by Education')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Inference:
print("1. Age: Most customers are between 40-60 years old.")
print("2. Income: Majority earn below $100k with some high outliers.")
print("3. Spending: Higher education levels correlate with higher spending.")
```
